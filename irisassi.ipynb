```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Classification',data=data)
plt.show()
```

```python
sns.scatterplot(x='SL',y='SW',hue='Classification',data=data)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

```python
sns.scatterplot(x='PL',y='PW',hue='Classification',data=data)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

```python
plt.hist(data['SL'],bins=5)
plt.title('sepal length')
plt.xlabel('sepal length')
plt.show()
```

```python
plt.hist(data['SW'],bins=5)
plt.title('sepal width')
plt.xlabel('sepal width')
plt.show()
```

```python
plt.hist(data['PL'],bins=5)
plt.title('petal length')
plt.xlabel('petal length')
plt.show()
```

```python
plt.hist(data['PW'],bins=5)
plt.title('petal length')
plt.xlabel('petal length')
plt.show()
```
